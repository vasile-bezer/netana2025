```python
from pathlib import Path
adjlist_path = Path().absolute().parent/'Data'/'course_network_l2.adjlist'
```

```python
import networkx as nx

G = nx.read_weighted_edgelist(adjlist_path)
nx.draw(G, with_labels=True, node_color='lightblue', node_size=1000, font_size=15)
```

```python

degrees = G.degree()

degrees_sorted = sorted(G.degree, key=lambda x: x[1], reverse=True)

print(degrees_sorted)
```

```python
G.edges()
```

```python
from matplotlib import pyplot as plt

node_degrees = dict(G.degree())

pos = nx.spring_layout(G)
edge_widths = [d["weight"]  for (u, v, d) in G.edges(data=True)]
normalized_weights = [1 + ((weight - min(edge_widths)) * (12 - 1)) / (max(edge_widths) - min(edge_widths)) for weight in edge_widths]


plt.figure(figsize=(10, 8), facecolor='white')
nx.draw_networkx_nodes(G, pos, node_size=[node_degrees[node] * 3 for node in G.nodes()], alpha=0.8)
nx.draw_networkx_edges(G, pos, width=normalized_weights, edge_color="grey", alpha=0.5)
nx.draw_networkx_labels(G, pos, {node: node if node in {node for node, degree in degrees_sorted[:3]} else "" for node in G.nodes()}, font_size=10, font_color="black")

plt.title("")
plt.axis("off")
plt.show()
```
