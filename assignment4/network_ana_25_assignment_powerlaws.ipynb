For this question, you'll be using NetworkX's implementation of the power law models discussed in the textbook -- specifically, the Barabasi-Albert graph generator.

This question asks you to show through simulations the degree distribution that is obtained from calls to **networkx.barabasi_albert_graph(n, m)** for varying values of parameter m, which controls the number of outbound edges from each node.

First, choose and fix a value for n, which controls the number of nodes in the generated network. Make sure that $n\ge{250}$.

Second, choose three values for m. For each of them, generate one network instance.

Finally, for each value of m, plot the degree distribution **in log-log scale** (x-axis: node degree, y-axis: number of nodes with given degree). Attach your plots as image files (please do not use the textbox).

#### Question 1.

Do the resulting degree distributions follow the shape anticipated by the discussion of Section 18.2 of the textbook? How does the value of m affect the distribution? Provide the answers to these questions in the textbox below.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1024
seed = 2
```

```python
G1 = nx.barabasi_albert_graph(n = N, m = 8, seed=seed)
G2 = nx.barabasi_albert_graph(n = N, m = 80, seed=seed)
G3 = nx.barabasi_albert_graph(n = N, m = 800, seed=seed)
```

```python
print("a)")
degree_sequence = sorted((d for n, d in G1.degree()), reverse=False)
dmax = max(degree_sequence)

fig = plt.figure("Degree of a random graph", figsize=(8, 8))
ax2 = fig.add_subplot()

#ax1 = fig.add_subplot(axgrid[3:, :2])
#ax1.plot(degree_sequence, "b-", marker="o")
#ax1.set_title("Degree Rank Plot")
#ax1.set_ylabel("Degree")
#ax1.set_xlabel("Rank")

ax2.bar(*np.unique(degree_sequence, return_counts=True), log=1)
ax2.set_xlabel("Degree")
ax2.set_ylabel("# of Nodes")

fig.tight_layout()
plt.title(f"Degree Distribution of Barabási-Albert Graph with n={N} m={8} in log scale", fontsize=12)
plt.grid(True)
plt.show()
```

```python
print("b)")
degree_sequence = sorted((d for n, d in G2.degree()), reverse=False)
dmax = max(degree_sequence)

fig = plt.figure("Degree of a random graph", figsize=(8, 8))

ax2 = fig.add_subplot()
ax2.bar(*np.unique(degree_sequence, return_counts=True), log=1)
ax2.set_xlabel("Degree")
ax2.set_ylabel("# of Nodes")

fig.tight_layout()
plt.title(f"Degree Distribution of Barabási-Albert Graph with n={N} m={80} in log scale", fontsize=12)
plt.grid(True)
plt.show()
```

```python
print("c)")
degree_sequence = sorted((d for n, d in G3.degree()), reverse=False)
dmax = max(degree_sequence)

fig = plt.figure("Degree of a random graph", figsize=(8, 8))

ax2 = fig.add_subplot()
ax2.bar(*np.unique(degree_sequence, return_counts=True), log=1)
ax2.set_xlabel("Degree")
ax2.set_ylabel("# of Nodes")

fig.tight_layout()
plt.title(f"Degree Distribution of Barabási-Albert Graph with n={N} m={800} in log scale", fontsize=12)
plt.grid(True)
plt.show()
```

The chosen values for n and m are respectively 1024 and (8, 80, 800). As for the first two plots (m=8, m=80) the shape of the distribution roughly follows the one of the power law although some values fall out of the ideal line (that is the power law distr.) but for the 3rd plot (m=800) the distribution appears to be divided in two populations, one behaving normal-like and the other uniform-like. In the 3rd case in particular it appears that a small amount of nodes have a much larger degree and the rest of the nodes remain with a much smaller degree. The heavy tailed behavior is displayed therefore the networkx.barabasi_albert_graph(n, m) algorithm does follow the expectations. 

As previously anticipated larger values of m make the distribution tend towards the right and with very large values the distribution allows for the Rich-Get-Richer scheme in full force, on the other hand with intermediate values the distribution shifts more and more to the right, trying to fit more and more into the power law distribution.
